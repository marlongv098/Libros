### **1) Funciones de velocidad y posición del cohete teniendo en cuenta La fuerza de gravedad (Fg) y la fuerza de resistencia (Fr)**

La ecuación diferencial que utilizaremos para el problema de propulsión de un cohete se obtiene mediante el siguiente proceso:


1. Empezamos a partir de la ecuación del momento. En este caso, como el cohete va quemando combustible a medida que va pasando el tiempo, la masa de este se ve disminuida.   

$$P(t)= m(t) * v(t)$$

Ahora describimos el cambio en el momento:



• Expansión del cambio de momento:  
Se descompone el cambio de momento considerando tanto la variación de la velocidad como la posible variación de la masa.


$$ΔP = P(t+Δt)-P(t)$$

$$ΔP=m(t+Δt)*v(t+Δt)-mv$$

$$ΔP=(m+Δm)*(v+Δv)-mv$$

$$ΔP=mv+mΔv+Δmv+ΔmΔv-mv$$

$$ΔP=mΔv+Δmv+ΔmΔv$$

Hasta el momento hemos obtenido una ecuación que describe cómo cambia el momento de un cohete cuando este no quema combustible, lo cual no tiene sentido cuando hablamos de un cohete en movimiento. Para solucionar lo anterior, al cambio de momento del cohete le debemos sumar el cambio del momento de los gases que este expulsa para obtener el cambio total.

$$ΔP_{gases} = -Δm*(v-c)$$

Como el gas va en direccion contraria al cohete, el cambio de momento de los gases es negativo.
Gracias a lo anterior obtenemos la siguiente ecuación que describe el cambio total del momento en el sistema.

$$ΔP_{total}= (mΔv + Δmv + ΔmΔv) +(-Δm)*(v-c)$$

Y a esta ecuación le aplicamos el siguiente procedimiento para llegar a la ecuación que buscamos:

$$ΔP= mΔv + Δmv + ΔmΔv -Δmv+ Δmc$$  

$$ΔP= mΔv + ΔmΔv + Δmc$$  

• División entre el intervalo de tiempo $Δt$:  
Para expresar la ecuación en términos de tasas de cambio, se divide todo entre $Δt$.  

$$\frac{ΔP}{Δt}=m *\frac{Δv}{Δt} + {Δm}*\frac{Δv}{Δt}+ \frac{Δm}{Δt}*c$$  

• Límite cuando $Δt$ tiende a $0$:  
Se toma el límite para obtener una ecuación diferencial. Como $ΔmΔv$ es un producto de cantidades diferenciales, su valor tiende a cero.  

$$\lim_{Δt \to0}{(\frac{ΔP}{Δt})}=m*~\lim_{Δt\to 0}{(\frac{Δv}{Δt})} + 0*~\lim_{Δt\to 0}{(\frac{Δv}{Δt})}+ c*\lim_{Δt\to 0}{(\frac{Δm}{Δt})}$$  


• Expresión en derivadas:  
Se obtiene la ecuación diferencial del momento en función del tiempo.  

$$\frac{dP}{dt}= m \frac{dv}{dt} + c\frac{dm}{dt}$$  

• Relación con la segunda ley de Newton:  
Dado que la fuerza $F$ es la derivada del momento respecto al tiempo, se llega a la ecuación general:  

$$F= m \frac{dv}{dt} + c\frac{dm}{dt}$$  

Sabemos que la fuerza neta es igual a la suma de todas las fuerzas que actuan sobre el objeto. En este caso, tomaremos en cuenta la fuerza gravitatoria la cual jala al cohete hacia abajo, o sea, es negativa; y la fuerza de resistencia que también va en sentido contrario al cohete.

$$F= F_G + F_R$$

$$F_G=-mg~~~~,~~~~~F_R=-kv$$

$$F= -mg -kv$$

$$~~~~~~~m \frac{dv}{dt} + c\frac{dm}{dt}= -mg-kv~~~~~~~~~~~(1)$$

Ahora debemos tener en cuenta la función de masa, la cual describe cómo cambia la masa del cohete mientras aumenta el tiempo y el combustible se consume a una tasa de quemado constante $-β$

$$m(0)=m_0~~~,~~~m(t)=m_0 -Bt~~~,~~~\frac{dm}{dt}=-B$$

Reemplazando $m(t)$ en $(1)$ obtenemos lo siguiente:

$$~~~~~~~~~~~(m_0-Bt)* \frac{dv}{dt}-cB=-(m_0 -Bt)g-kv~~~~~~~~~~~(2)$$

La ecuación $(2)$  es la ecuación diferencial que hemos estado buscando y es la que nos ayudará a encontrar la función de velocidad y posición de un cohete en propulsión. A continuación realizamos el proceso para resolver la ecuación diferencial y encontrar las funciones antes mencionadas.

***

2) Para resolver esta ecuación diferencial utilizaremos el factor integrante. Primero debemos organizar la ecuación para que nos quede organizada de manera similar a la siguiente expresión.

$$\frac{dy}{dx}+P(x)*y=Q(x)$$

Organizamos:

• Ecuación diferencial original:  
Partimos de una ecuación diferencial que modela la variación de la velocidad en función del tiempo, considerando un sistema con pérdida de masa y fuerzas externas.  

$$(m_0-Bt)* \frac{dv}{dt}-cB=-(m_0 -Bt)g-kv$$  

$$(m_0-Bt)* \frac{dv}{dt}+kv=Bc-(m_0 -Bt)g$$  

$$\frac{(m_0 -Bt)\frac{dv}{dt}}{(m_0 -Bt)}+ \frac{kv}{(m_0-Bt)}= \frac{Bc-(m_0-Bt)g}{(m_0-Bt)}$$  

$$\frac{dv}{dt}+ \frac{k}{(m_0-Bt)}v= \frac{Bc}{(m_0-Bt)}-\frac{(m_0-Bt)}{(m_0-Bt)}g$$  

• Forma final de la ecuación diferencial:  
Se simplifica la expresión obtenida en la forma estándar de ecuación diferencial de primer orden.  

$$~~~~~~~~~~~\frac{dv}{dt}+ \frac{k}{(m_0-Bt)}v= \frac{Bc}{(m_0-Bt)}-g~~~~~~~~~~~(3)$$ 


Ya que la hemos organizado, ahora podemos definir nuestra $P(t)$ y encontrar nuestro factor Integral: $M$

$$\frac{dv}{dt}+ P(t)v=Q(t)$$

$$P(t)=\frac{k}{(m_0-Bt)}$$

• Se encuentra el factor integrante:

$$M=e^{∫P(t)\,dt}$$

$$M=e^{∫\frac{k}{(m_0-Bt)}dt}$$

• Se realiza un cambio de variable:

$$u=m_0-Bt~~~~~~,~~~~~-\frac{du}{B}=dt$$

• Se transforma la integral usando el cambio de variable:

$$-k*\frac{1}{B}∫\frac{du}{u}$$

$$-k*\frac{1}{B}~Ln|u|$$

• Se reemplaza por el valor original:

$$M=e^{\frac{-k}{B}Ln|m_0-Bt|}$$

• Se simplifica usando la propiedad de los logaritmos:

$$M=|m_0-Bt|^{\frac{-k}{B}}$$


Ahora multiplicamos $(3)$ por el factor integral que acabamos de encontrar.

$$M*\frac{dv}{dt}+ \frac{k}{(m_0-Bt)}Mv= M\frac{Bc}{(m_0-Bt)}-gM$$

$$(m_0-Bt)^{\frac{-k}{B}}*\frac{dv}{dt}+ \frac{k}{(m_0-Bt)}(m_0-Bt)^{\frac{-k}{B}}v= (m_0-Bt)^{\frac{-k}{B}}\frac{Bc}{(m_0-Bt)}-g(m_0-Bt)^{\frac{-k}{B}}$$

Al lado izquierdo tenemos la derivada del producto $M*v$, por lo tanto, la ecuación nos queda asi:

$$\frac{d(vM)}{dt}=(m_0-Bt)^{\frac{-k}{B}}*\frac{Bc}{(m_0-Bt)}-g(m_0-Bt)^{\frac{-k}{B}}$$

Ahora, integramos a ambos lados para encontrar la función de velocidad.

$$v*(m_0-Bt)^{\frac{-k}{B}}=∫[(m_0-Bt)^{\frac{-k}{B}}*\frac{Bc}{(m_0-Bt)}-g(m_0-Bt)^{\frac{-k}{B}}]dt$$


$$v(m_0-Bt)^{\frac{-k}{B}}=∫[(m_0-Bt)^{\frac{-k}{B}} * \frac{Bc}{(m_0-Bt)}]dt~~- g∫(m_0-Bt)^{\frac{-k}{B}} dt$$

• Se reorganiza la fracción del primer término:

$$v(m_0-Bt)^{\frac{-k}{B}}=∫[ \frac{(m_0-Bt)^{\frac{-k}{B}} * Bc}{(m_0-Bt)}]dt~~- g∫(m_0-Bt)^{\frac{-k}{B}} dt$$

• Se reescriben los exponentes de manera más clara:

$$v(m_0-Bt)^{\frac{-k}{B}}=∫[ (m_0-Bt)^{\frac{-k}{B}} * Bc~(m_0-Bt)^{-1}]dt~~- g∫(m_0-Bt)^{\frac{-k}{B}} dt$$


$$v(m_0-Bt)^{\frac{-k}{B}}=∫[ (m_0-Bt)^{\frac{-k}{B}-1} * Bc]dt~~- g∫(m_0-Bt)^{\frac{-k}{B}} dt$$

$$v(m_0-Bt)^{\frac{-k}{B}}=Bc∫(m_0-Bt)^{\frac{-k}{B}-1} dt~~- g∫(m_0-Bt)^{\frac{-k}{B}} dt$$

• Se realiza un cambio de variable para facilitar la integración:

$$u=(m_0-Bt)~~~,~~~du=-B~dt~~~,~~~\frac{du}{-B}=dt$$

$$v(m_0-Bt)^{\frac{-k}{B}}=Bc∫(u)^{\frac{-k}{B}-1} *\frac{du}{-B}~~- g∫u^{\frac{-k}{B}} \frac{du}{-B}$$

$$v(m_0-Bt)^{\frac{-k}{B}}=\frac{Bc}{-B}∫(u)^{\frac{-k}{B}-1} du~~+\frac{1}{B}~ g∫u^{\frac{-k}{B}} du$$

$$v(m_0-Bt)^{\frac{-k}{B}}=-c∫(u)^{\frac{-k}{B}-1} du~~+\frac{1}{B}~ g∫u^{\frac{-k}{B}} du$$

• Se resuelven las integrales aplicando la regla de potencias:

$$vM=-c \frac{u^{\frac{-k}{B}}}{\frac{-k}{B}}+\frac{1}{B}~g*\frac{u^{\frac{-k}{B}+1}}{\frac{-k}{B}+1}+W$$

• Se reorganizan los términos para simplificar la expresión:

$$vM=cB \frac{u^{\frac{-k}{B}}}{k}+\frac{u^{\frac{-k}{B}+1}}{-k+B}~~g+W$$

$$v=\frac{cB~ \frac{u^{\frac{-k}{B}}}{k}+\frac{u^{\frac{-k}{B}+1}}{-k+B}~~g+W}{(m_0-Bt)^{\frac{-k}{B}}}$$

• Se descompone la fracción:

$$v=\frac{\frac{cB~(m_0-Bt)^{\frac{-k}{B}}}{k}}{(m_0-Bt)^{\frac{-k}{B}}}+\frac{\frac{(m_0-Bt)^{\frac{-k}{B}+1}g}{-k+B}}{(m_0-Bt)^{\frac{-k}{B}}}+\frac{W}{(m_0-Bt)^{\frac{-k}{B}}}$$

$$v=\frac{cB(m_0 - Bt)^{\frac{-k}{B}}}{k(m_0-Bt)^{\frac{-k}{B}}}+\frac{(m_0-Bt)^{\frac{-k}{B}+1}g}{(m_0-Bt)^{\frac{-k}{B}}(-k+B)}+\frac{W}{(m_0-Bt)^{\frac{-k}{B}}}$$

• Se cancelan exponentes en los términos correspondientes:

$$v=\frac{cB(m_0-Bt)^{\frac{-k}{B}}}{k(m_0-Bt)^{\frac{-k}{B}}}+\frac{(m_0-Bt)*g}{(-k+B)}+ \frac{W}{(m_0-Bt)^{\frac{-k}{B}}}$$


• Expresión final de la velocidad en función del tiempo:

$$v(t)=\frac{cB}{k}+\frac{(m_0 -Bt)~g}{-k+B} + \frac{W}{(m_0-Bt)^{\frac{-k}{B}}}$$


Gracias a todo el proceso anterior, hemos encontrado nuestra función de velocidad teniendo en cuenta la fuerza de gravedad y de resistencia:

**Función de velocidad teniendo en cuenta Fg y Fr**

$$v(t)=\frac{g}{(-k+B)} * (m_0-Bt)+ \frac{Bc}{k} + \frac{W}{(m_0-Bt)^{\frac{-k}{B}}}~~~~~~~~~~~~~~~~~$$  

3) Integrando la funcion de velocidad $v$ podemos encontrar nuestra función de posición $y$:

$$dy=[\frac{g}{(-k+B)} * (m_0-Bt)+ \frac{Bc}{k} + \frac{W}{(m_0-Bt)^{\frac{-k}{B}}}]~dt$$

$$y=\frac{g}{(-k+B)} ∫(m_0 -Bt) dt~~+ \frac{Bc}{k}∫dt~~+W∫\frac{1}{(m_0-Bt)^{\frac{-k}{B}}}dt$$

$$z=m_0-Bt,~~~~~~~~\frac{-dz}{B}=dt$$

$$y=-\frac{g}{(-k+B)B} ∫(z) dz~~+ \frac{Bc}{k}∫dt~~-\frac{W}{B}∫\frac{1}{(z)^{\frac{-k}{B}}}dz$$

$$y=-\frac{g}{(-k+B)B} * \frac{(m_0 -Bt)^{2}}{2}~~+ \frac{Bc}{k}t~~-\frac{W}{B} ~~\frac{z^{\frac{k}{B}+1}}{\frac{k+B}{B}}+G$$

Por último, hemos obtenido nuestra función de posición $y$ y hemos completado nuestro objetivo.

**Funcion de posición teniendo en cuenta Fg y Fr**

$$y(t)=\frac{Bc}{k}t~~-\frac{g}{(-k+B)B} *\frac{(m_0-Bt)^{2}}{2}~~-W ~~\frac{(m_o-Bt)^{\frac{k}{B}+1}}{k+B}+G$$





### **2) Funciones de velocidad y posición del cohete sin tener en cuenta la fuerza de resistencia (Fr)**

El cambio del momento total del cohete consiste en la suma del momento del cohete y el momento que genera la expulsión de gases, siendo variable la masa. Así, tenemos que:

$$\frac{dP}{dt}= F= m \frac{dv}{dt} + c * \frac{dm}{dt}~~~~~~~~~~~~~~~~~~~~~~~~(1)$$

Ahora, si consideramos que el combustible se consume a una tasa de quemado constante tenemos:

$$\frac{dm}{dt}= -α ~~~~~,~~~~α>0$$

Por lo tanto, la función de la masa es:

$$m(t)= m_0 -αt$$

Por lo que $(1)$ nos queda:

$$F=(m_0 - αt)~\frac{dv}{dt}-c~α~~~~~~~(2)$$

Como no vamos a considerar la fuerza de resistencia del aire, tenemos que la fuerza es igual a:

$$F=F_R+F_G$$

$$F=-mg-kv~~~,~~~k=0$$

$$F=-mg$$

Por lo tanto, $(2)$ nos queda:

$$(m_0-α t)~\frac{dv}{dt}-c~α= - (m_0-α~t)g~~~~~~~~~~~~~~~~~~(3)$$

• Se despeja el término que contiene la derivada:
$$(m_0-α t)~\frac{dv}{dt}=c~α - (m_0-α~t)g$$


$$\frac{dv}{dt}=\frac{c~α}{m_0-α~t}-g$$

• Se reescribe en forma diferencial:
$$dv=[\frac{c~α}{m_0-α~t}-g]*dt$$


Ahora integramos para encontrar la función velocidad respecto al tiempo $v(t)$, lo que nos da como resultado lo siguiente:

$$v(t)=-c*Ln|m_0- αt|-gt+W~~~~~~~~~~~~(4)$$

Reescribimos la anterior función de la siguiente manera para encontrar nuestra función de posición.

Sabemos que $~~v(0)=v_0~~$. Reemplazando en la función $v(t):$

$$v_0=-c*Ln|m_0- α~(0)|-g(0)+W$$

$$v_0=-c*Ln|m_0|+W$$

$$v_0+cLn|m_0|=W$$

Reemplazando el valor de $~~W~~$ en $~~(4)~~$ tenemos lo siguiente:

$$v(t)=-c*Ln|m_0-α~t|-gt+v_0+c*Ln|m_0|$$

$$v(t)=v_0-g*t+c*ln(m_0)-c*ln(m_o-αt)$$

$$v(t)=v_0-g*t+c*[ln(m_0)-ln(m_o-αt)]$$

**Función de velocidad sin tener en cuenta la fuerza de resistencia $(Fr)$**

$$v(t)=v_0-g*t+c*ln(\frac{m_0}{m_o-αt})~~~~~~~~~~~(5)$$

Ahora integremos $(5)$ para encontrar nuestra función de posición.

$$dy=[~v_0-g*t+c*ln(\frac{m_0}{m_o-αt})~]*dt$$

$$y=v_0∫dt-g∫t*dt+∫c*ln(\frac{m_0}{m_o-αt})~dt$$

$$y=v_0∫dt-g∫t*dt+I_1~~~~~~~~~~(6)$$

Procedemos a resolver $I_1$:

$$I_1=∫c*ln(\frac{m_0}{m_o-αt})~dt$$

Hacemos un cambio de variable para facilitar el proceso:

$$z=m_0-αt~~~~,~~~~ \frac{-dz}{α}=dt$$

$$I_1=\frac{-c}{α}∫[~ln~(m_0)-ln(z)~]~dz$$

$$I_1=\frac{-c}{α}*(∫ln~(m_0)~dz-∫ ln(z)~dz)$$

$$I_1=\frac{-c}{α}*ln~(m_0)∫dz+\frac{c}{α}∫ ln(z)*dz$$

$$I_1=\frac{-c}{α}*ln~(m_0)∫dz+\frac{c}{α}*I_2~~~~~~~~~~~~~(7)$$

Resolvamos $~~I_2~~$:

$$I_2=∫ ln(z)*dz$$

Utilizamos integración por partes: $~~∫udv=uv-∫vdu~~$

$$u=ln(z)~~~~~⟹~~~~du=\frac{1}{z}~dz$$

$$dv=dz~~~~~~⟹~~~~~~~v=z$$

$$I_2=∫ ln(z)*dz=ln(z)*z-∫z*\frac{1}{z}*dz$$

$$I_2=ln(z)*z-z$$

Reemplazamos $I_2$ en $(7)$:

$$I_1=\frac{-c}{α}*ln~(m_0)∫dz+\frac{c}{α}*(~ln(z)*z-z~)$$

$$I_1=\frac{-c}{α}*ln~(m_0)~z+\frac{c}{α}*ln(z)*z-\frac{c}{α}*z$$

Y reemplazando $I_1$ en $(6)$:

$$y=v_0∫dt-g∫t*dt-\frac{c}{α}*ln~(m_0)~z+\frac{c}{α}*ln(z)*z-\frac{c}{α}*z$$

$$y=v_0*t-\frac{g}{2}*t^{2}-\frac{c}{α}*ln~(m_0)~z+\frac{c}{α}*ln(z)*z-\frac{c}{α}*z$$

$$y=v_0~t-\frac{g}{2}*t^{2}-\frac{c*z}{α}*(~ln(m_0)-ln(z)+1~)$$

$$y=v_0~t-\frac{g}{2}*t^{2}-\frac{c*z}{α}*(~ln(\frac{m_0}{z})+1~)$$

$$y=v_0~t-\frac{g}{2}*t^{2}-\frac{c*z}{α}*ln(\frac{m_0}{z})-\frac{c*z}{α}$$

$$y=v_0~t-\frac{g}{2}*t^{2}-\frac{c}{α}*ln(\frac{m_0}{m_0-αt})*(m_0-αt)-\frac{c}{α}*(m_0-αt)$$

Ahora hacemos la siguiente suma para simplificar la expresión:

$$v_0~t-\frac{c}{α}*(m_0-αt)$$

$$= v_0*t-\frac{c}{α}*m_0+\frac{c}{α}*α*t$$

$$= t*(v_0+c)-\frac{c}{α}*m_0$$

Entonces la función de posición nos queda de la siguiente manera:

**Función de posición sin tener en cuenta la fuerza de resistencia $(Fr)$**

$$y=(~v_0+c~)*t-\frac{g}{2}*t^{2}-\frac{c}{α}*ln(\frac{m_0}{m_0-αt})*(m_0-αt)+G$$





### **3. ¿Que sucede si el cohete se acelera en el espacio libre, donde no hay gravedad ni resistencia?**

En el espacio libre, donde no hay fuerzas externas como la gravedad o la resistencia del aire, el movimiento del cohete sigue las leyes de Newton. Si el cohete se acelera, esto significa que está expulsando masa en forma de gas a alta velocidad.

Si se detiene la propulsión, el cohete continuará moviéndose a velocidad constante en línea recta, ya que no hay fuerzas externas que alteren su estado de movimiento.

Sin embargo, antes de que se acabe el combustible el cambio en la velocidad se comporta de acuerdo a la ecuación que se obtiene a continuación:

• Se parte de la ecuación de la segunda ley de Newton aplicada a una masa variable:

$$m \frac{dv}{dt}=-c\frac{dm}{dt}$$

• Se reorganizan los términos para separar las variables:

$$m~dv~dt=-c~dm~dt$$

• Se integra ambos lados:

$$\int m~dv~dt=∫-c~dm~dt$$

• Se eliminan los diferenciales $dt$, dejando una ecuación separable:

$$m~dv=-c~dm$$

• Se despeja \( dv \):

$$dv = \frac{-c}{m}~dm$$

• Se integran ambos lados entre los valores iniciales y finales:

$$\int_{v1}^{v0} dv = -c \int_{m1}^{m0} \frac{1}{m}~dm$$

• Se resuelve la integral del lado derecho:

$$v1-v0=-c~(ln(m_1)-ln(m_0))$$

• Se aplica la propiedad del logaritmo:

$$\Delta v=c~(ln(m_0)-ln(m_1))$$

• Se expresa la variación de velocidad en su forma final:

$$\Delta v=c~ln(\frac{m_0}{m_1})$$

Masa inicial del cohete con combustible :

* Para resolver esta pregunta hay que utilizar la función de velocidad sin tener en cuenta $Fr$ la cual es:

$$v(t)=-c*Ln|m_0-α~t|-gt+W~~~~~~~~~~~~~~~~~~~~(1)$$

Además, asumiendo que $~~v_0=0~~$ cuando $~~t=0~~$, obtenemos:

$$v_0=-c*Ln|m_0- α~(0)|-g(0)+W$$

$$v_0=-c*Ln|m_0|+W$$

$$v_0+cLn|m_0|=W$$

Reemplazando el valor de $~~W~~$ en $~~(1)~~$ tenemos lo siguiente:

$$v(t)=-c*Ln|m_0-α~t|-gt+v_0+c*Ln|m_0|$$

$$v(t)=v_0-g*t+c*ln(m_0)-c*ln(m_o-αt)$$

$$v(t)=v_0-g*t+c*[ln(m_0)-ln(m_o-αt)]$$

$$v(t)=v_0-g*t+c*ln(\frac{m_0}{m_o-αt})~~~~~~~~~~~(2)$$

*Sustituimos los datos que nos da el problema en $~~(2)~~$ para encontrar la velocidad del cohete a los 30 segundos. Asumiendo que $~~v_0=0~~$ nos queda:

$$v(t)=-9.81*t+2500*ln(\frac{5000}{5000-50t})$$

Evaluamos en $~~t=30~~$:

$$v(30)=-9.81*(30)+2500*ln(\frac{5000}{5000-50*(30)})$$

$$v(30)=-294,3+891,687$$

$$v(30)≈597,387$$

$R//$ La velocidad del cohete a los $~~30~~$ segundos es de $~~597,387~~m/s~~$




```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
c = 2500
g = 9.81
m0 = 5000
alpha = 50
t_max = 80
n_particiones = 50
dt = t_max / n_particiones

# Función de la ecuación diferencial dv/dt = f(t, v)
def ecuacion_diferencial(t, v):
    m = m0 - alpha * t
    if m > 1000:
        return (c * alpha) / m - g
    else:
        return -g

# Solución exacta de la velocidad basada en la ecuación transcrita
def solucion_real(t):
    m = m0 - alpha * t
    return -g * t + c * np.log(abs(m0 / m))

# Método de Euler para velocidad
def euler(f, t0, v0, t_final, n):
    t_values = np.linspace(t0, t_final, n + 1)
    v_values = np.zeros(n + 1)
    v_values[0] = v0

    for i in range(n):
        v_values[i + 1] = v_values[i] + f(t_values[i], v_values[i]) * dt

    return t_values, v_values

# Condición inicial
v0 = 0

# Resolver usando Euler
t_values, v_euler = euler(ecuacion_diferencial, 0, v0, t_max, n_particiones)

# Calcular solución exacta
v_real = np.array([solucion_real(t) for t in t_values])

# Evaluar en t = 30 y t = 80
v_t30_euler = np.interp(30, t_values, v_euler)
v_t30_real = solucion_real(30)
v_t80_euler = np.interp(80, t_values, v_euler)
v_t80_real = solucion_real(80)

print(f"Velocidad en t=30 (Euler): {v_t30_euler:.2f} m/s")
print(f"Velocidad en t=30 (Exacta): {v_t30_real:.2f} m/s")
#print(f"Velocidad en t=80 (Euler): {v_t80_euler:.2f} m/s")
#print(f"Velocidad en t=80 (Exacta): {v_t80_real:.2f} m/s")

# Graficar resultados
plt.figure(figsize=(10, 5))
plt.plot(t_values, v_real, label="Solución Exacta", color="red")
plt.plot(t_values, v_euler, "-", label="Euler (Aproximación)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.title("Velocidad del cohete")
plt.legend()
plt.grid()
plt.show()
```

Gracias a la gráfica podemos observar que a medida que pasa el tiempo la velocidad del cohete aumenta lo cual puede deberse a que mientras menos masa haya, gracias a la perdida de combustible, la aceleración aumenta, lo que hace que la velocidad crezca rápidamente.

Masa inicial del cohete con combustible :

$$m_0=5000~kg$$

Masa final del cohete sin combustible:

$$m_1=1000~~kg$$

Tasa de quemado del combustible:

$$α=50~~kg/s$$

Si $~~t_1~~$ es el tiempo en que el cohete alcanza su masa final sin combustible y la tasa de quemado del combustible es igual a:

$$\frac{dm}{dt}=-α$$

entonces tenemos que:

$$\int_{m_0}^{m_1}dm=\int_{0}^{t1}-α*dt$$

$$m_1-m_0=-α~*t_1 + α*(0)$$

$$m_1=m_0-α*t_1$$

Reemplazando valores:

$$1000=5000-50*t_1$$

$$t_1=\frac{5000-1000}{50}$$

$$t_1=80$$

$R//~~$ El cohete tarda $~~80~s~$ en consumir todo su combustible.



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
m0 = 5000    # Masa inicial (kg)
alpha = 50   # Tasa de pérdida de masa (kg/s)
t0 = 0       # Tiempo inicial
dt = 1       # Paso de tiempo
t_max = 100  # Tiempo máximo de simulación

# Método de Euler para la pérdida de masa
def euler_mass(m0, alpha, t0, dt, t_max):
    t_values = [t0]
    m_values = [m0]

    t = t0
    m = m0

    while m > 1000 and t < t_max:
        m = m - alpha * dt
        t += dt
        t_values.append(t)
        m_values.append(m)

    return t_values, m_values

# Resolver usando Euler
t_values, m_values = euler_mass(m0, alpha, t0, dt, t_max)

# Encontrar el tiempo donde m ≈ 1000 kg
t_1000 = None
for i in range(len(m_values)):
    if m_values[i] <= 1000:
        t_1000 = t_values[i]
        break

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(t_values, m_values, label="Masa del cohete", color="blue")
plt.axhline(y=1000, color="red", linestyle="--", label="Masa = 1000 kg")
plt.axvline(x=t_1000, color="green", linestyle="--", label=f"t ≈ {t_1000} s")
plt.scatter([t_1000], [1000], color="black", zorder=3, label="Punto crítico")

# Configurar etiquetas
plt.xlabel("Tiempo (s)")
plt.ylabel("Masa (kg)")
plt.title("Disminución de la masa del cohete")
plt.legend()
plt.grid()
plt.show()

# Imprimir el resultado
print(f"La masa alcanza 1000 kg en t ≈ {t_1000} segundos")

```

Gracias a la gráfica podemos observar el comportamiento de la masa durante el transcurso del tiempo, pudiendo notar que a los 80 segundos el cohete ha gastado todo su combustible, lo cual es un punto clave en el analisis del vuelo de este, ya que después de este tiempo la velocidad del cohete empieza a cambiar y la unica fuerza que actua sobre él es la de gravedad lo que provoca una desaceleración que va subiendo con el tiempo.

* **Datos del problema.**

Masa inicial del cohete con combustible :

$$m_0=5000~kg$$

Velocidad constante a la que salen los gases del cohete:

$$c=2500~m/s$$

Tasa de quemado del combustible:

$$α=50~kg/s$$

Tiempo en que el cohete se queda sin combustible:

$$t_1=80~~s$$

Masa final del cohete sin combustible:

$$m_1=1000~~kg$$

Gravedad:

$$g=9.81~~m/s^{2}$$

Si evaluamos $~~t_1~~$ en la función de velocidad sin tener en cuenta la resistencia que encontramos en el punto a), obtendremos la velocidad que llevaba en ese momento:

$$v(t)=v_0-g*t+c*ln(\frac{m_0}{m_o-αt})$$

$$v(t)=-9.81*t+2500*ln(\frac{5000}{5000-50t})$$

$$v(t_1)=-9.81*(t_1)+2500*ln(\frac{5000}{m_1})$$

$$v(80)=-9.81*(80)+2500*ln(\frac{5000}{1000})$$

$$v(80)=-784.8+4023.594$$

$$v(80)≈3238.794$$

$R//~~$ A los $~~80~~$ segundos, momento en que el cohete se queda sin combustible, la velocidad de este será de $~~3238.794~~m/s~~$



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
c = 2500
g = 9.81
m0 = 5000
alpha = 50
t_max = 80
n_particiones = 50
dt = t_max / n_particiones

# Función de la ecuación diferencial dv/dt = f(t, v)
def ecuacion_diferencial(t, v):
    m = m0 - alpha * t
    if m > 1000:
        return (c * alpha) / m - g
    else:
        return -g

# Solución exacta de la velocidad basada en la ecuación transcrita
def solucion_real(t):
    m = m0 - alpha * t
    return -g * t + c * np.log(abs(m0 / m))

# Método de Euler para velocidad
def euler(f, t0, v0, t_final, n):
    t_values = np.linspace(t0, t_final, n + 1)
    v_values = np.zeros(n + 1)
    v_values[0] = v0

    for i in range(n):
        v_values[i + 1] = v_values[i] + f(t_values[i], v_values[i]) * dt

    return t_values, v_values

# Condición inicial
v0 = 0

# Resolver usando Euler
t_values, v_euler = euler(ecuacion_diferencial, 0, v0, t_max, n_particiones)

# Calcular solución exacta
v_real = np.array([solucion_real(t) for t in t_values])

# Evaluar en t = 30 y t = 80
v_t30_euler = np.interp(30, t_values, v_euler)
v_t30_real = solucion_real(30)
v_t80_euler = np.interp(80, t_values, v_euler)
v_t80_real = solucion_real(80)

#print(f"Velocidad en t=30 (Euler): {v_t30_euler:.2f} m/s")
#print(f"Velocidad en t=30 (Exacta): {v_t30_real:.2f} m/s")
print(f"Velocidad en t=80 (Euler): {v_t80_euler:.2f} m/s")
print(f"Velocidad en t=80 (Exacta): {v_t80_real:.2f} m/s")

# Graficar resultados
plt.figure(figsize=(10, 5))
plt.plot(t_values, v_real, label="Solución Exacta", color="red")
plt.plot(t_values, v_euler, "-", label="Euler (Aproximación)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.title("Velocidad del cohete")
plt.legend()
plt.grid()
plt.show()
```

* **Datos del problema.**

Masa inicial del cohete con combustible :

$$m_0=5000~kg$$

Velocidad constante a la que salen los gases del cohete:

$$c=2500~m/s$$

Tasa de quemado del combustible:

$$α=50~kg/s$$

Tiempo:

$$t=30~~s$$

Velocidad inicial:
$$v_0=0$$

Gravedad:

$$g=9.81~~m/s^{2}$$

Para resolver esta pregunta necesitamos utilizar la función de posición cuando la fuerza de resistencia es despreciable:

$$y(t)=(~v_0+c~)*t-\frac{g}{2}*t^{2}-\frac{c}{α}*ln(\frac{m_0}{m_0-αt})*(m_0-αt)+G$$

Utilizamos la condición inicial $~~y(0)=0~~$ para encontrar el valor de la constante de integración.

$$0=(~v_0+c~)*(0)-\frac{g}{2}*(0)^{2}-\frac{c}{α}*ln(\frac{m_0}{m_0})*(m_0)+G$$

$$0=G$$

Ahora reemplazamos los datos que tenemos en la función de posición:

$$y(t)=2500*t-\frac{9.81}{2}*t^{2}-\frac{2500}{50}*ln(\frac{5000}{5000-50t})*(5000-50t)$$

Evaluamos la función en $t=30$:

$$y(30)=2500*(30)-\frac{9.81}{2}*(30)^{2}-50*ln(\frac{5000}{5000-50*(30)})*(~5000-50*(30)~)$$

$$y(30)=75000-4414.5-62418.11519$$

$$y(30)≈ 8167.38481$$

$R//$ La altura del cohete después de $~~30~s~~$ de ascenso es de $~~8167.38481~$ metros.



```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
c = 2500      # Velocidad de eyección (m/s)
g = 9.81      # Gravedad (m/s²)
m0 = 5000     # Masa inicial (kg)
alpha = 50    # Consumo de combustible (kg/s)
t_max = 80    # Tiempo máximo en segundos (combustible se agota en 80s)
n_particiones = 100  # Número de particiones
dt = t_max / n_particiones  # Paso de tiempo

# Función de la ecuación diferencial dy/dt = v
def ecuacion_diferencial(t, y, v):
    return v

# Función para la velocidad V(t)
def velocidad(t):
    if t <= 80:
        m = m0 - alpha * t
        return c * np.log(m0 / m) - g * t
    else:
        v80 = c * np.log(m0 / 1000) - g * 80
        return v80 - g * (t - 80)

# Función de la solución exacta de la posición
def solucion_real(t):
    if t <= 80:
        return (c * t) - (0.5 * g * t**2) - ((c / alpha) * np.log(m0 / (m0 - alpha * t))) * (m0 - alpha * t)
    else:
        y80 = (c * 80) - (0.5 * g * 80**2) - ((c / alpha) * np.log(m0 / 1000)) * 1000
        v80 = velocidad(80)
        return y80 + v80 * (t - 80) - 0.5 * g * (t - 80)**2

# Método de Euler para y(t)
def euler(f, t0, y0, v0, t_final, n):
    t_values = np.linspace(t0, t_final, n + 1)  # Vector de tiempos
    y_values = np.zeros(n + 1)  # Almacena valores de posición
    v_values = np.zeros(n + 1)  # Almacena valores de velocidad
    y_values[0] = y0  # Condición inicial de posición
    v_values[0] = v0  # Condición inicial de velocidad

    for i in range(n):
        v_values[i + 1] = velocidad(t_values[i])
        y_values[i + 1] = y_values[i] + f(t_values[i], y_values[i], v_values[i]) * dt

    return t_values, y_values

# Condiciones iniciales
y0 = 0  # Posición inicial
t_values, y_euler = euler(ecuacion_diferencial, 0, y0, velocidad(0), t_max, n_particiones)

# Calcular solución exacta
y_real = np.array([solucion_real(t) for t in t_values])

# Imprimir valores en t=80 y t=30
t_test = [30, 80]
for t in t_test:
    print(f"t = {t}s -> Solución exacta: y({t}) = {solucion_real(t)} m, Euler: y({t}) = {np.interp(t, t_values, y_euler)} m")

# Graficar resultados
plt.figure(figsize=(18, 5))
plt.plot(t_values, y_real, label="Solución Exacta", color="red")
plt.plot(t_values, y_euler, "-", label="Euler (Aproximación)")  # Puntos con línea
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.title("Comparación entre Euler y la solución real para la posición")
plt.legend()
plt.grid()
plt.yticks(np.arange(0, max(y_real) + 10000, 10000))  # Escala de 10,000 en 10,000
plt.show()
```

Con el paso del tiempo el cohete va alcanzando más altura gracias a la gran velocidad que lleva. En el momento en que el cohete gasta todo su combustible, la unica fuerza que va actuar sobre este será la de gravedad, porque ya no existe esa fuerza de propulsión dada por el motor del cohete, lo cual va a provocar que después de los 80 segundos este empiece a perder velocidad y por lo tanto pierda altura.
